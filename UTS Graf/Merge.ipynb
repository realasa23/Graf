```python
import pandas as pd
```

```python

# 1. Load data
df_wiki = pd.read_csv('wikidata.csv')
df_db = pd.read_csv('dbpedia.csv')

#  2. CLEANING DBPEDIA
def clean_dbpedia_uri(uri):
    if pd.isna(uri): return uri
    # Ambil teks setelah '/' terakhir dan ganti '_' jadi spasi
    name = str(uri).split('/')[-1].replace('_', ' ')
    return name.lower().strip()

# Terapkan pembersihan ke kolom-kolom DBpedia
df_db['book_clean'] = df_db['book'].apply(clean_dbpedia_uri)
df_db['publisher_clean'] = df_db['publisher'].apply(clean_dbpedia_uri)
# Ambil tahunnya saja dari publishDate (misal: '1997-06-26' -> '1997')
df_db['release_year'] = df_db['publishDate'].str[:4]

#  3. NORMALISASI WIKIDATA
df_wiki['book_clean'] = df_wiki['bookLabel'].str.lower().str.strip()
df_wiki['publisher_clean'] = df_wiki['publisherLabel'].str.lower().str.strip()

#  4. MERGING
df_final = pd.merge(
    df_wiki,
    df_db[['book_clean', 'publisher_clean', 'release_year']],
    on=['book_clean', 'publisher_clean'],
    how='outer'
)

#  5. RAPIKAN HASIL AKHIR
df_final['final_book_title'] = df_final['bookLabel'].fillna(df_final['book_clean'])
df_final['final_publisher'] = df_final['publisherLabel'].fillna(df_final['publisher_clean'])

cols_to_save = [
    'final_book_title', 'authorLabel', 'genreLabel',
    'final_publisher', 'release_year'
]
df_final_save = df_final[cols_to_save]

# Simpan ke CSV final
df_final_save.to_csv('data_literatur_final.csv', index=False)
```
